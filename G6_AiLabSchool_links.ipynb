````python
# decorador que no afecta funciones
def decorador_nulo(func):
  return func


def saludo():
  return "hola!"

greet = decorador_nulo(saludo)

greet()
````

````python
@decorador_nulo
def saludo2():
  return "hola mundo!"

saludo2()
````

````python
def uppercase(func):
  def wrapper():
    resultado_original = func()
    resultado_modificado = resultado_original.upper()
    return resultado_modificado
  return wrapper


  
````

````python
@uppercase
def greet():
  return "hello!"


````

````python
@uppercase
def greeting2():
  return "este texto es minuscula"

````

````python
greeting2()
````
